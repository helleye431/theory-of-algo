## Задание 17

### Текст задания

Ваши друзья открывают компанию, занимающуюся компьютерной безопасностью, и им необходимо получить лицензии на `n` разных криптографических продуктов. Согласно требованиям, они могут получать не более одной лицензии в месяц.

Каждая лицензия сейчас стоит `$100`, но со временем цена растет по экспоненциальному закону: лицензия `j` каждый месяц умножается на коэффициент `r_j > 1`. Если лицензия `j` приобретается через `t` месяцев, она будет стоить `100 * r_j^t`. Все коэффициенты различны.

Нужно определить порядок покупки лицензий так, чтобы общая потраченная сумма была минимальной. Требуется алгоритм, работающий за полиномиальное время.

Идея:

Если какая-то лицензия дорожает быстрее других, ее лучше купить раньше.  
Значит, выгоднее всего покупать лицензии по убыванию коэффициента роста: сначала самые быстро дорожающие, потом остальные.


```python
rates = [1.05, 1.20, 1.10, 1.03]

order = list(range(len(rates)))
order.sort(key=lambda i: rates[i], reverse=True)

total_cost = 0
month = 0

while month < len(order):
    index = order[month]
    total_cost += 100 * (rates[index] ** month)
    month += 1

order, total_cost
```

## Задание 17(2)

### Текст задания

Теперь вместо покупки лицензий рассматривается продажа оборудования, цена которого со временем уменьшается.

Цена элемента `i` уменьшается с коэффициентом `r_i < 1` в месяц, начиная с `$100`, так что при продаже через `t` месяцев за него можно получить `100 * r_i^t`.

Если в месяц продается только один вид оборудования, нужно определить оптимальный порядок продажи.


Идея: 
Тут цена со временем падает.  
Если какой-то предмет теряет цену медленнее, его можно подержать дольше. А тот, который дешевеет быстрее, лучше продать раньше.

В итоге снова получается порядок по убыванию коэффициента `r`: сначала большие `r`, потом меньшие.


```python
rates = [0.95, 0.80, 0.90, 0.70]

order = list(range(len(rates)))
order.sort(key=lambda i: rates[i], reverse=True)

total_revenue = 0
month = 0

while month < len(order):
    index = order[month]
    total_revenue += 100 * (rates[index] ** month)
    month += 1

order, total_revenue
```

## Задание 18

### Текст задания

Дана последовательность событий `S` длины `n` и короткая последовательность `S'` длины `m`. Нужно определить, является ли `S'` подпоследовательностью `S`, то есть можно ли удалить из `S` некоторые элементы так, чтобы оставшиеся элементы в том же порядке совпали с `S'`.

Требуется алгоритм со временем работы `O(m + n)`.

идея:

Просто идем слева направо по большой последовательности.  
Держим указатель на текущий элемент в короткой последовательности.  
Если нашли совпадение, двигаем указатель дальше.  
Если в конце удалось пройти всю короткую последовательность, значит это подпоследовательность.


```python

S = ["купить Yahoo", "купить eBay", "купить Yahoo", "купить Oracle"]
s = [
    "купить Amazon",
    "купить Yahoo",
    "купить eBay",
    "купить Yahoo",
    "купить Yahoo",
    "купить Oracle"
]

i = 0
j = 0

while i < len(S) and j < len(s):
    if S[i] == s[j]:
        i += 1
    j += 1

result = i == len(S)
result
```
